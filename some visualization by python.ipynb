```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define data
data = {
    "Model": ["Decision Tree"] * 4 + ["Random Forest"] * 4 + ["XGBoost"] * 4 + ["GBM"] * 4,
    "Variant": ["Baseline", "Baseline", "SMOTE", "SMOTE"] * 4,
    "Threshold": [0.5000, 0.0938, 0.5000, 0.3660, 
                  0.5000, 0.0450, 0.5000, 0.3450, 
                  0.5000, 0.0853, 0.5000, 0.2614, 
                  0.5000, 0.0852, 0.5000, 0.2921],
    "Sensitivity": [0.0444, 0.7502, 0.6264, 0.7701, 
                    0.0222, 0.7643, 0.5802, 0.7913, 
                    0.0554, 0.8487, 0.4640, 0.8140, 
                    0.0520, 0.8284, 0.4945, 0.8051]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Pivot the DataFrame for Baseline and SMOTE variants
df_baseline = df[df["Variant"] == "Baseline"].pivot(index="Model", columns="Threshold", values="Sensitivity")
df_smote = df[df["Variant"] == "SMOTE"].pivot(index="Model", columns="Threshold", values="Sensitivity")

# Plot heatmaps
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

colormap = sns.color_palette("coolwarm", as_cmap=True)

# Baseline heatmap
sns.heatmap(df_baseline, cmap=colormap, annot=True, fmt=".4f", linewidths=0.5, ax=axes[0], 
            center=0.5, cbar_kws={"shrink": 0.8})
axes[0].set_title("Sensitivity Heatmap (Baseline)")

# SMOTE heatmap
sns.heatmap(df_smote, cmap=colormap, annot=True, fmt=".4f", linewidths=0.5, ax=axes[1], 
            center=0.5, cbar_kws={"shrink": 0.8})
axes[1].set_title("Sensitivity Heatmap (SMOTE)")

plt.tight_layout()
plt.show()
```
